# üç± The DataFrame: A Collection of Series
A DataFrame is essentially a vertical stack of Series objects that share the same Index.

* The Series (1D): A single column of data.

* The DataFrame (2D): A collection of these columns joined together to form a table. 

```python
import pandas as pd 
import numpy as np 
```

```python
data = {
    'Name':  ["K M Tala","Mahin", "Hasib","Toma"],
    'Age': [20, 19, 28, 21],
    'City': ['Barguna', 'patuakhali', 'Dhaka', 'Kulana'],
    'Salary': [100000, 70000, 62000, 55000]
}

df = pd.DataFrame(data)
```

```python
df
```

 ### you also can create  dataframe using list  

```python
data_list = [
    ['K M Talha', 20, 'Barguna', 100000],
    ['Mahin', 19, 'patuakhali', 70000],
    ['Hasib', 27, 'Dhaka', 62000],
    ['Toma', 21, 'Kulana', 55000]
]

#df2 =pd.DataFrame(data_list)
#df2
columns = ["Name","Age","City","Salary"]  # if we not use colums then it give me index like 0,1,2 on columns
df2 = pd.DataFrame(data_list,columns=columns)  
df2
```

## Selection and Indexing of Columns

```python
df2['Age']  # select singal column 
```

```python
df2[["Name","City"]] # to select multipal columns 
```

## Creating a new column

```python
df2["Designation"] = ["Eng.","Doct.","Teacher","Eng"]
df2
```

 ## Removing Columns
 

```python
df2.drop("Name", axis=1) #select asis=0 if you want to remove from row and  or select asix=1 to remove from column. 
```

```python
 #Use inplace if you want to permanently delete something from memory. 
# if you want to drop multipal column for drop then use multiple column  list 

df2.drop("Designation",axis=1, inplace= True)
```
