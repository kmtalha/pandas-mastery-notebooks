# Pandas Series

A Series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index
.

### There are multiple ways to create a pandas Series:
EX: using array,dic,list
* we can also remove the indexes if we dont want using  index = False 

```python
import pandas as pd 
import numpy as np 
```

```python
labales = [1,2,3,4,5]
friend_list = ["K M Tala","Mahin", "Hasib","Toma","Radia"]
dictonary =  {"Name": "Roll",
             "K M Talha": 111,
             "Mahin":222,
             "Hasib":333,
             "Toma":444,
             "Radia":555}
arr = np.array([10,20,30,40,50])


```

```python
pd.Series(friend_list)
```

```python
pd.Series(friend_list,index = labales)
```

```python
pd.Series(arr)
```

```python
pd.Series(dictonary)
```
